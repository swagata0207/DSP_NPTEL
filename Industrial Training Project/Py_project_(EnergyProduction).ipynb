```
# Train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df1[['Year']], df1[['EnergyIndex']], train_size = 0.8, test_size = 0.2, random_state = 100)
```

```
lr = linear_model.LinearRegression()
lr.fit(X_train, y_train)
```

```
X_train.head()
```

```
y_train.head()
```

```
m = lr.coef_

c = lr.intercept_

print('Slope:', m[0])
print('Intercept:', c)
```

```
predicted_data = lr.predict(X_test[['Year']])
predicted_data[0:5]
```

```
plt.figure(figsize = (10,6))
plt.scatter(X_test, y_test)
plt.plot(X_test['Year'], predicted_data, 'r')
plt.xticks(np.arange(1969, 1990, 5.0))
plt.title('Best fit line for test dataset')
plt.xlabel('Year')
plt.ylabel('EnergyIndex')
plt.show()
```

```
# Calculation of R squared
from sklearn.metrics import r2_score
r2 = r2_score(test_data_array, predicted_data)
print(r2)
```
