```
# importing python libraries NumPy, pandas, Matplotlib and seaborn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# to suppress warnings
import warnings
warnings.filterwarnings('ignore')
```

```
# to read the datafile(in csv format) and import 'DATE' Column into pandas dataframe as Datetime object
ep = pd.read_csv('EnergyProduction.csv',parse_dates = ['DATE'],index_col = 'DATE')
```

```
plt.figure(figsize=(15,6))
plt.plot(ep.index, ep['EnergyIndex'])
plt.xlabel('Year'), plt.ylabel('Energy Index'), plt.title('Energy Production')
```

```
plt.figure(figsize=(15,6))
plt.scatter(ep.index, ep['EnergyIndex'], color = 'purple')
plt.xlabel('Year'), plt.ylabel('Energy Index'), plt.title('Energy Production')
```

```
plt.figure(figsize=(15,6))
plt.bar(ep.index, ep['EnergyIndex'], edgecolor = 'purple')
plt.xlabel('Year'), plt.ylabel('Energy Index'), plt.title('Energy Production')
```

```
plt.figure(figsize=(15,6))
sns.set_style('whitegrid')
sns.distplot(ep['EnergyIndex'],bins= 80,hist_kws={'edgecolor':'black'})
plt.title('Energy Index densities'), plt.xlabel('Energy Index'), plt.ylabel('Density')
```
